```python
import pandas as pd
import pandas_datareader as pdr
df = pdr.get_data_yahoo('AAPL',start='2012-1-1',end='2016-1-1')
```

```python
df=df['Close']
```

```python
returns = df.pct_change()[1:]
```

```python
returns.iplot(kind='hist')
```

```python
returns.iplot(kind='line')
```

```python
sma = df.rolling(30).mean()
```

```python
both = pd.concat([df,sma],axis=1)
```

```python
both.columns = ['Close','Moving']
```

```python
both.head()
```

```python
both.iplot(kind='line',)
```

```python
both1=both.resample("M").mean()
```

```python
both1.iplot(kind='line')
```

```python
test = pdr.get_data_yahoo(['SPY', 'AAPL'],start='2012-1-1',end='2016-1-1')
```

```python
test=test['Close']
```

```python
test.head()
```

```python
import plotly.figure_factory as ff
```

```python
bri = pdr.get_data_yahoo(['MCD','MSFT'], start='2012-1-1', end='2016-1-1')
bri=bri['Close']
```

```python
df1 = pd.concat([test,bri],axis=1)
```

```python
df1.head()
```

```python
figure = ff.create_scatterplotmatrix(df1[['AAPL','SPY','MCD','MSFT']],diag='histogram')
```

```python
figure.show()
```

```python
corr=df1.corr()
```

```python
figure = ff.create_annotated_heatmap(z=corr.values,x=list(corr.columns),y=list(corr.index),annotation_text=corr.round(2).values,showscale=True)
```

```python
figure.show()
```
