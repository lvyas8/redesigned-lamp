```python
election = pd.read_csv('C:\\Users\\LajpatraiVyas\\Desktop\\Manipulating_Df\\pennsylvania2012_turnout.csv')
red_blue = {'Obama':'blue', 'Romney':'red'}
election['win_col'] = election['winner'].map(red_blue)
election.head()
```

```python
from scipy.stats import zscore
turnout_zscore = zscore(election['turnout'])
election['turnout_zscore']=turnout_zscore
election.head()
```
