```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # importing necessary packages 
```

```python
x = np.random.rand(50) #creates a data with 50 numbers between 0 and 1
```

```python
y = x*2 + np.random.normal(0,0.1,50) # adding some noise to the data
```

```python
np.cov(x,y)[0,1] # using numpy covariance function to calculate covariance
```

```python
X = np.random.rand(50) #creating another set of 50 numbers
Y = x + 4 # adding some noise to the data
print("covariance of  X & Y \n" + str(np.cov(X,Y))) # printing the covariance matrix. becareful while reading the matrix
print("correlation of X & Y \n" + str(np.corrcoef(X,Y))) # printing the correlation matrix again be careful while reading
```

```python
cov_mat = np.cov(X,Y)
```

```python
error = cov_mat[0,0] - X.var(ddof=1)
print('error:' + str(error))
```

```python
plt.scatter(X,Y) #using scatterplot function to visualize the data and we can see the noise
plt.xlabel('X Values') # adding x-axis label to visualization
plt.ylabel('Y Values') # adding y-axis label to visualization
print('Manually calc Correlation:',str(np.cov(X,Y)[0,1]/(np.std(X)*np.std(Y)))) # calculating correlation manually
print('Built in Correlation', str(np.corrcoef(X,Y)[0,1])); # calculating correlation using the built in function
```

```python
a=np.random.rand(50) # creating a set of 50 random numbers
b=a + np.random.normal(0,0.1,50) #by using normally distributed number we generate a high correlation
print('Very high correlation:',str(np.corrcoef(a,b)[0,1])) # using built in correlation function to calculate correlation
plt.scatter(a,b) # syntax for scatter plot
plt.xlabel('a values') # adding labels to x-axis
plt.ylabel('b values'); # adding labels to y-axis
```

```python
i = np.random.rand(50)
j = i + np.random.normal(0,0.2,50)
print('relatively low correlation:', str(np.corrcoef(i,j)[0,1]))
plt.scatter(i,j)
plt.xlabel('i values')
plt.ylabel('j values');
```

```python
o = np.random.rand(50)
p = -o + np.random.normal(0,0.2,50) #creating inverse relation and that is why it is left to right sloping downwards
print('relatively low correlation:', str(np.corrcoef(o,p)[0,1]))
plt.scatter(o,p)
plt.xlabel('o values')
plt.ylabel('p values');
```
